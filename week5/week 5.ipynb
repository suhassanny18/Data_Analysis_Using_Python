```python
import matplotlib.pyplot as plt
import numpy as np 
marks=[90,50,40,60,55,44,30,10,35,84]
grade_intervals=[0,35,50,60,70,80,90,100]
plt.title('student grade analysis')
plt.hist(marks,bins=30,histtype='bar',rwidth=0.7,facecolor='g')
plt.xticks=(grade_intervals)
plt.xlabel=('distribution of students')
plt.ylabel=('count of students')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 
marks=[90,50,40,60,55,44,30,10,35,84]
grade_intervals=[0,35,50,60,70,80,90,100]
plt.title('student grade analysis')
plt.hist(marks,grade_intervals,histtype='stepfilled',rwidth=0.7,facecolor='g')
plt.xticks=(grade_intervals)
plt.xlabel=('distribution of students')
plt.ylabel=('count of students')
plt.show()
```
